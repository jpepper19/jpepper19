# Example Code - From Thesis Census Data

```python
import pandas as pd
import numpy as np
import os
```

```python
def load_income_data(folder_path):
    dataframes = {}
    
    for year in range(2013, 2024):  # Loop from 2014 to 2023
        filename = f"{year}_Income_B19001.csv"
        file_path = os.path.join(folder_path, filename)
        
        if os.path.exists(file_path):  # Check if file exists
            df = pd.read_csv(file_path)
            df.index = range(1, len(df) + 1)  # Assign numerical index starting at 1
            dataframes[f"{year}_B19001"] = df
        else:
            print(f"Warning: {filename} not found.")
    
    return dataframes

# Test Run
folder_path = "C:/Users/jepep/OneDrive/UTK/Thesis/Datasets/ACS_INCOME_B19001"  
dataframes = load_income_data(folder_path)

df_2014 = dataframes["2014_B19001"]
df_2014.head(18)
```
